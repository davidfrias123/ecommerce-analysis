```python

# Imports and Data Loading
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Set display settings
pd.set_option('display.max_columns', None)

# Load merged dataset
merged_df = pd.read_csv("merged_ecommerce_data.csv")

print("✅ Data loaded successfully!")
merged_df.head()
```

```python
# Data Cleaning Summary

# Basic info
merged_df.info()

# Missing values
merged_df.isna().sum().sort_values(ascending=False).head(15)

```

```python
# Descriptive Statistics

merged_df.describe(include='all').transpose()
```

```python
# Key Visuals (Matplotlib + Seaborn)

# Interaction type distribution
sns.countplot(data=merged_df, x='interaction_type',
              order=merged_df['interaction_type'].value_counts().index)
plt.title("Distribution of Interaction Types")
plt.show()

# Gender distribution
sns.countplot(data=merged_df, x='gender')
plt.title("Customer Gender Distribution")
plt.show()

# Age distribution
sns.histplot(data=merged_df, x='age', bins=20, kde=True)
plt.title("Age Distribution of Customers")
plt.show()

# Purchases by gender
sns.countplot(data=merged_df[merged_df['interaction_type']=='purchase'], x='gender')
plt.title("Purchases by Gender")
plt.show()
```

```python
# Interactive Visuals (Plotly)

# Interactive: Purchases by Age
purchase_data = merged_df[merged_df['interaction_type']=='purchase']
fig = px.histogram(purchase_data, x='age', color='gender',
                   nbins=20, title="Purchases by Age and Gender")
fig.show()

# Interactive: Payment method
fig = px.bar(merged_df, y='payment_method', color='interaction_type',
             title="Interactions by Payment Method", orientation='h')
fig.show()
```

```python
# Insights & Observations (Markdown)
### Key Insights

- Most interactions are *views*, but only a smaller portion are *purchases*.
- Male and female customers are nearly balanced, but one group might have slightly higher purchase conversion.
- Customers aged 25–40 are the most active buyers.
- PayPal and Credit Card are the top payment methods.
```
